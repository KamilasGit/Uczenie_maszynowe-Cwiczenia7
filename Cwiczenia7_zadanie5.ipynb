```python
# Tic Tac Toe

# Create the board
board = [' ' for _ in range(9)]

# Define the winning combinations
winning_combinations = [
    (0, 1, 2), (3, 4, 5), (6, 7, 8),  # rows
    (0, 3, 6), (1, 4, 7), (2, 5, 8),  # columns
    (0, 4, 8), (2, 4, 6)  # diagonals
]

# Function to display the board
def display_board():
    print('-------------')
    for i in range(0, 9, 3):
        print(f'| {board[i]} | {board[i+1]} | {board[i+2]} |')
        print('-------------')

# Function to check if someone has won
def check_winner(symbol):
    for combination in winning_combinations:
        if board[combination[0]] == board[combination[1]] == board[combination[2]] == symbol:
            return True
    return False

# Function to play the game
def play_game():
    print("Let's start a game!")
    current_symbol = 'X'
    game_over = False

    while not game_over:
        display_board()
        print(f"Player {current_symbol}'s turn")
        position = int(input("Enter a position (First row: 1-3, second row: 4-6, third row: 7-9): ")) - 1

        if board[position] == ' ':
            board[position] = current_symbol
            if check_winner(current_symbol):
                display_board()
                print(f'Player {current_symbol} wins!')
                game_over = True
            elif ' ' not in board:
                display_board()
                print("It's a tie!")
                game_over = True
            else:
                current_symbol = 'O' if current_symbol == 'X' else 'X'
        else:
            print('Invalid move. Try again.')

# Start the game
play_game()
```
